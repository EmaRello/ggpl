Definisco le funzioni utili alla generazione della vegetazione:
- createTree(treeType,rTree,hTree,hTrunk) : genera un albero dati il tipo, il raggio, l'altezza complessiva della pianta e quella del tronco (posta a zero per i cespugli)
- rowOfTrees (treeTypes,rTree,hTreeMin,hTreeMax,hTrunkMin,hTrunkMax,orientation,numberOfTrees) : restituisce un filare di alberi. Vengono passati come parametri una lista di tipologie di alberi da includere, le caratteristiche degli alberi (raggio ed altezze massime-minime), l'orientamento della fila ed il numero di esemplari di cui è composta. Gli alberi hanno caratteristiche random all'interno dei valori passati come parametro e vengono randomicamente disallineati.
- groupOfTrees (treeTypes,rTree,hTreeMin,hTreeMax,hTrunkMin,hTrunkMax,rGroup,numberOfTrees) : restituisce analogamente alla funzione precedente un gruppo di alberi in un'area (raggio di tale area è un parametro). All'interno dell'area vengono posizionati alberi secondo una distribuzione randomica (coordinate polari random ) omogenea (distribuzione di Archimede)

```python
def createTree(treeType,rTree,hTree,hTrunk):
    """
    Funzione che genera un albero
    
    Args:
        treeType(int): tipologia di albero desiderata
        rTree(float): raggio dell'albero
        hTree(float): altezza dell'albero
        hTrunk(float): altezza del tronco
        
    Returns:
        tree(oggetto hpc): albero generato
    """
    
    # definisco le textures necessarie
    textureExtended = 'textures/foliage2.jpg'
    textureNormal = 'textures/foliage.jpg'
    
    # gestisco alberi di tipo 3
    if treeType == 3 :
        textureExtended = 'textures/flower.jpg'
        treeType = 2
    
    # gestisco alberi di tipo 1
    hFoliage = hTree-hTrunk
    if treeType == 1 :
        foliage = MKPOL([[[0,0,0],[0,rTree,0],[0,rTree*1.5/4,hFoliage*1/3],
                         [0,rTree*3/4,hFoliage*1/3],[0,rTree*1/4,hFoliage*2/3],[0,rTree*2/4,hFoliage*2/3],
                         [0,0,hFoliage],[0,0,hFoliage*2/3],[0,0,hFoliage*1/3]],[[1,2,3,9],[9,4,5,8],[8,6,7]],1])
        foliage = TEXTURE(textureExtended)(foliage)
        circle1 = TEXTURE(textureNormal)(CYLINDER([rTree*1/2,0])(10))
        circle2 = TEXTURE(textureNormal)(CYLINDER([rTree*3/8,0])(10))
        circle3 = TEXTURE(textureNormal)(CYLINDER([rTree*2/8,0])(10))
        foliage = STRUCT([foliage,circle1,T(3)(hFoliage*1/3),circle2,T(3)(hFoliage*1/3),circle3])
    
    # gestisco alberi di tipo 2
    elif treeType == 2 :
        foliage = T(3)(rTree*1/2)(R([1,3])(math.pi/2)(CYLINDER([rTree*1/2,0])(20)))
        foliage = S([2,3])([2,hFoliage/rTree])(foliage)
        foliage = TEXTURE(textureExtended)(foliage)
        circle1 = TEXTURE(textureExtended)(CYLINDER([rTree*1/2,0])(10))
        circle2 = TEXTURE(textureExtended)(CYLINDER([rTree*5/6,0])(10))
        circle3 = TEXTURE(textureExtended)(CYLINDER([rTree*1/2,0])(10))
        foliage = STRUCT([foliage,T(3)(hFoliage*1/4),circle1,T(3)(hFoliage*1/4),circle2,T(3)(hFoliage*1/4),circle3])

    # conferisco volume alla figura
    foliage = STRUCT([foliage,S(2)(-1),foliage])
    i = 0
    while i < 3.14 :
        i += 3.14/4
        foliage = STRUCT([foliage,R([1,2])(i),foliage])
    
    # assemblo tronco e chioma
    trunk = MATERIAL(woodMaterial)(CYLINDER([rTree/16,hTrunk])(10))
    tree = STRUCT([trunk,T(3)(hTrunk),foliage])
    
    # ritorno l'albero
    return tree

#tree = createTree(1,3.,6.,1.)
#VIEW(tree)

def rowOfTrees (treeTypes,rTree,hTreeMin,hTreeMax,hTrunkMin,hTrunkMax,orientation,numberOfTrees) :
    """
    Funzione che genera una fila di alberi
    
    Args:
        treeTypes(List[int]): tipologie di alberi da includere
        rTree(float): raggio dell'albero
        hTreeMin(float): altezza minima dell'albero
        hTreeMax(float): altezza massima dell'albero
        hTrunkMin(float): altezza minima del tronco
        hTrunkMax(float): altezza massima del tronco
        orientation(int): indica l'orientamento (verticale/orizzontale) della fila
        numberOfTrees(int) : numero di alberi da includere
        
    Returns:
        row(oggetto hpc): fila di alberi generata
    """
    
    # genero il primo albero
    hTree = (random.random())*(hTreeMax-hTreeMin)+hTreeMin
    hTrunk =(random.random())*(hTrunkMax-hTrunkMin)+hTrunkMin
    tree = createTree(treeTypes[0],rTree,hTree,hTrunk)
    row = T(1)(rTree)(tree)
    
    # dispongo randomicamente gli altri alberi
    for i in range (1,numberOfTrees) :
        hTree = (random.random())*(hTreeMax-hTreeMin)+hTreeMin
        hTrunk = (random.random())*(hTrunkMax-hTrunkMin)+hTrunkMin
        yTree = ((random.random())*rTree/2)-rTree/4
        treeTypeIndex = int((random.random())*float(len(treeTypes))-000000.1)
        tree = createTree(treeTypes[treeTypeIndex],rTree,hTree,hTrunk)
        row = STRUCT([row,T([1,2])([rTree+i*2*rTree+((random.random())*rTree/2),yTree]),tree])
    
    # ruoto la fila secondo l' orientamento
    row = R([1,2])(orientation*math.pi/2)(row)
    
    # ritorno la fila
    return row

def groupOfTrees (treeTypes,rTree,hTreeMin,hTreeMax,hTrunkMin,hTrunkMax,rGroup,numberOfTrees) :
    """
    Funzione che genera un gruppo di alberi
    
    Args:
        treeTypes(List[int]): tipologie di alberi da includere
        rTree(float): raggio dell'albero
        hTreeMin(float): altezza minima dell'albero
        hTreeMax(float): altezza massima dell'albero
        hTrunkMin(float): altezza minima del tronco
        hTrunkMax(float): altezza massima del tronco
        rGroup(float): raggio dell' area da popolare
        numberOfTrees(int) : numero di alberi da includere
        
    Returns:
        group(oggetto hpc): fila di alberi generata
    """
    # genero il primo albero
    hTree = (random.random())*(hTreeMax-hTreeMin)+hTreeMin
    hTrunk =(random.random())*(hTrunkMax-hTrunkMin)+hTrunkMin
    tree = createTree(treeTypes[0],rTree,hTree,hTrunk)
    group = tree
    
    # genero e posiziono randomicamente gli altri alberi
    for i in range (1,numberOfTrees) :
        hTree = (random.random())*(hTreeMax-hTreeMin)+hTreeMin
        hTrunk = (random.random())*(hTrunkMax-hTrunkMin)+hTrunkMin
        treeTypeIndex = int((random.random())*float(len(treeTypes))-000000.1)
        
        archimedesRandomicNumber = random.random()+random.random()
        if archimedesRandomicNumber > 1 : archimedesRandomicNumber = 2-archimedesRandomicNumber
        radiusTree = archimedesRandomicNumber*(rGroup-rTree)
        
        angleTree = (random.random())*(math.pi*2)
        tree = createTree(treeTypes[treeTypeIndex],rTree,hTree,hTrunk)
        tree = T(1)(radiusTree)(tree)
        tree = R([1,2])(angleTree)(tree)
        group = STRUCT([group,tree])
    
    # predispongo un bordo di controllo 
    #group = STRUCT([group,COLOR(RED)(SKEL_1(CYLINDER([rGroup,hTreeMax])(20)))])
    
    #ritorno il gruppo
    return group


#VIEW(rowOfTrees([1,2],3.,6.,8.,.8,1.,0,5))
#VIEW(groupOfTrees([1,2],3.,6.,8.,.8,1.,10,10))
```

Definisco la funzione utile a generare un ponte date le dimensioni rispettivamente sugli assi X, Y del ponte totale e sull'asse Y del traforo.

```python
def createBridge(depth,length,holeLength) :
    """
    Funzione che genera un ponte
    
    Args:
        depth(float): dimensione totale del ponte su asse X
        length(float): dimensione totale del ponte su asse Y
        holeLength(float): dimensione del traforo su asse Y
        
    Returns:
        bridge(oggetto hpc): ponte generato
    """
    
    # predispongo le parti del ponte
    length = length/2
    holeLength = holeLength/2
    
    bridgeLength = holeLength+(length-holeLength)/16
    
    hole = CYLINDER([holeLength,depth])(50)
    
    bridge = CYLINDER([bridgeLength,depth*0.9])(50)
    protection = CYLINDER([length,depth*0.05])(50)
    controlCube = T([1,2,3])([-length,-length,-depth])(CUBOID([length*2,length,depth*2]))
    
    # ricavo il ponte
    bridge = DIFFERENCE([bridge,hole,controlCube])
    bridge = T(3)(-depth*0.9/2)(bridge)
    bridge = R([2,3])(math.pi/2)(bridge)
    bridge = R([1,2])(math.pi/2)(bridge)
    
    # ricavo la protezione laterale
    protection = DIFFERENCE([protection,hole,controlCube])
    protection = T(3)(-depth*0.05/2)(protection)
    protection = R([2,3])(math.pi/2)(protection)
    protection = R([1,2])(math.pi/2)(protection)
    
    bridge = STRUCT([bridge,T(1)(depth*0.95/2),protection])
    bridge = STRUCT([bridge,T(1)(-depth*0.95/2),protection])
    
    bridge = S(3)(0.5)(bridge)
    
    #genero le scalinate (workshop_03)
    stairs = adaptedStairs (depth*0.9,bridgeLength,bridgeLength/2)
    stairs = T([1,2])([-depth*0.9/2,-length])(stairs)
    stairs = STRUCT([stairs,S(2)(-1),stairs])
    
    bridge = STRUCT([bridge,stairs])
    
    #genero la passerella
    riser = getStepDimensions(depth*0.9,bridgeLength/2,bridgeLength/2)[0]
    catwalk = T([1,2,3])([-depth*0.9/2,-(2*length-2*bridgeLength)/2,bridgeLength/2-riser])(CUBOID([depth*0.9,2*length-2*bridgeLength,riser]))
    
    bridge = STRUCT([bridge,catwalk])
    
    # applico colore materiale al ponte
    bridge = MATERIAL(woodMaterial)(bridge)
    
    # dispongo bordi di controllo e ritorno il ponte
    
    #edge = COLOR(RED)(SKEL_1(T([1,2])([-depth/2,-length])(CUBOID([depth,length*2,length/2]))))
    #bridge = STRUCT([bridge,edge])
    
    return bridge  
```

Definisco la funzioni utile a generare le parti di un lampione di dimensioni predefinite.

```python
def createLamp () :
    """
    Funzione che genera la lanterna del lampione
    
    Args:
    
    Returns:
        lamp(oggetto hpc): lanterna generata
    """
    lamp = CUBOID([0.2,0.2,0.3])
    lamp = SKEL_1(lamp)
    lamp = T([1,2])([-0.2/2,-0.2/2])(lamp)
    lamp = OFFSET([0.04,0.04,0.04])(lamp)
    lamp = T([1,2])([-0.02,-0.02])(lamp)
    bottom = T([1,2])([-.08,-.08])(CUBOID([.16,.16,0.04]))
    lamp = STRUCT([lamp,bottom])
    top = MKPOL([[[-0.24/2,-0.24/2,0.34],[0.24/2,-0.24/2,0.34],[0.24/2,0.24/2,0.34],[-0.24/2,0.24/2,0.34],[0,0,0.44]],[[1,2,3,4,5]],1])
    lamp = STRUCT([lamp,top])
    lamp = MATERIAL(obsidianMaterial)(lamp)
    glass = MATERIAL(glassMaterial)(T([1,2,3])([-0.08,-0.08,0.04])(CUBOID([.16,.16,0.26])))
    lamp = STRUCT([lamp,glass])
    #lamp = STRUCT([lamp,T([1,2])([-0.24/2,-0.24/2]),COLOR(RED)(SKEL_1(CUBOID([0.24,0.24,0.34])))])

    return lamp

def createStreetLamp () :
    """
    Funzione che genera il lampione
    
    Args:
    
    Returns:
        streetLamp(oggetto hpc): lampione generato
    """
    
    # genero ed assemblo il tronco ed i bracci del lampione
    trunkPoints = [[.05,1.7],[.05,2.2]]
    trunk = Plasm.power(BEZIERSTRIPE([trunkPoints,.1,20]),QUOTE([0]))
    branchRightPoints = [[.5,2.],[.5,2.2]]
    branchRight = Plasm.power(BEZIERSTRIPE([branchRightPoints,.08,20]),QUOTE([0]))
    junctionRightPoints = [[.05,1.7],[.5,1.7],[.05,2.],[.5,2.]]
    junctionRight = Plasm.power(BEZIERSTRIPE([junctionRightPoints,.08,20]),QUOTE([0]))

    streetLamp = STRUCT([trunk,branchRight,junctionRight])
    streetLamp = STRUCT([streetLamp,S(1)(-1),streetLamp])

    streetLamp = OFFSET([0,0,.1])(streetLamp)
    streetLamp = T([2,3])([-1.1,-0.05])(streetLamp)
    streetLamp = R([2,3])(math.pi/2)(streetLamp)
    streetLamp = T([3])([1.1])(streetLamp)
    streetLamp = STRUCT([streetLamp,R([1,2])(math.pi/2),streetLamp])
    streetLamp = STRUCT([streetLamp,CYLINDER([0.1,1.7])(20)])
    
    #genero ed assemblo la base
    streetLamp = STRUCT([T([1,2])([-0.2,-0.2]),CUBOID([0.4,0.4,0.2]),T([1,2,3])([0.2,0.2,0.2]),CYLINDER([0.18,0.1])(20),T(3)(0.1),streetLamp])
    streetLamp = MATERIAL(obsidianMaterial)(streetLamp)

    #assemblo i lumi al corpo del lampione
    lamp = createLamp ()

    streetLamp = STRUCT([streetLamp,T(3)(2.5),lamp])
    lamp = S([1,2,3])([.8,.8,.8])(lamp)
    streetLamp = STRUCT([streetLamp,T([1,2,3])([.5,0,2.5]),lamp])
    streetLamp = STRUCT([streetLamp,T([1,2,3])([0,.5,2.5]),lamp])
    streetLamp = STRUCT([streetLamp,T([1,2,3])([-.5,0,2.5]),lamp])
    streetLamp = STRUCT([streetLamp,T([1,2,3])([0,-.5,2.5]),lamp])

    #ritorno il lampione
    return streetLamp
```

Definisco la finzione che genera una strada. Per una presentazione realistica la strada viene suddivisa, vengono trattate separatamente le sue parti ed infine viene riassemblata.

```python
def createStreet (bezierPoints,depth,intervals,dz,controlCubes) :
    """
    Funzione che genera una strada
    
    Args:
        bezierPoints(List[List[float]]): punti e tensori che definiscono la curva
        depth(float): larghezza della strada
        intervals(int): intervalli di definizione della curva (precisione)
        dz(float): altezza della strada
        controlCubes(oggetto hpc): eventuali aree in cui la strada deve essere tagliata
    Returns:
        street(oggetto hpc): strada generata
    """
    
    #genero le varie "strisce" che compongono la strada
    street100 = Plasm.power(BEZIERSTRIPE([bezierPoints,depth,intervals]),QUOTE([dz]))
    street90 = Plasm.power(BEZIERSTRIPE([bezierPoints,depth*0.90,intervals]),QUOTE([dz]))
    street86 = Plasm.power(BEZIERSTRIPE([bezierPoints,depth*0.86,intervals]),QUOTE([dz]))
    street52 = Plasm.power(BEZIERSTRIPE([bezierPoints,depth*0.52,intervals]),QUOTE([dz]))
    street48 = Plasm.power(BEZIERSTRIPE([bezierPoints,depth*0.48,intervals]),QUOTE([dz]))
    street14 = Plasm.power(BEZIERSTRIPE([bezierPoints,depth*0.14,intervals]),QUOTE([dz]))
    street10 = Plasm.power(BEZIERSTRIPE([bezierPoints,depth*0.10,intervals]),QUOTE([dz]))
    
    # applico colori materiali alle varie "strisce"
    asphaltExt = MATERIAL(asphaltMaterial)(DIFFERENCE([street100,street90]+(controlCubes)))
    lineExt = MATERIAL(whiteMaterial)(DIFFERENCE([street90,street86]+(controlCubes)))
    asphaltMidExt = MATERIAL(asphaltMaterial)(DIFFERENCE([street86,street52]+(controlCubes)))
    lineMid = MATERIAL(whiteMaterial)(DIFFERENCE([street52,street48]+(controlCubes)))
    asphaltMidInt = MATERIAL(asphaltMaterial)(DIFFERENCE([street48,street14]+(controlCubes)))
    lineInt = MATERIAL(whiteMaterial)(DIFFERENCE([street14,street10]+(controlCubes)))
    asphaltInt = MATERIAL(asphaltMaterial)(DIFFERENCE([street10]+(controlCubes)))
    
    street = STRUCT([asphaltExt,lineExt,asphaltMidExt,lineMid,asphaltMidInt,lineInt,asphaltInt]) 
    return street

#VIEW(createStreet(punti,3.0,30,0.01))
```

Analogamente definisco una funzione che genera sentieri/camminamenti

```python
def createWalkway (bezierPoints,depth,intervals,dz,controlCubes) :
    """
    Funzione che genera un camminamento
    
    Args:
        bezierPoints(List[List[float]]): punti e tensori che definiscono la curva
        depth(float): larghezza del camminamento
        intervals(int): intervalli di definizione della curva (precisione)
        dz(float): altezza del camminamento
        controlCubes(oggetto hpc): eventuali aree in cui il camminamento deve essere tagliata
    Returns:
        walkway(oggetto hpc): camminamento generato
    """
    walkway = Plasm.power(BEZIERSTRIPE([bezierPoints,depth,intervals]),QUOTE([dz]))
    walkway = DIFFERENCE([walkway]+controlCubes)
    walkway = MATERIAL(sandMaterial)(walkway)
    return walkway
```

Utilizzo la funzione precedente per definirne un' altra che genera i sentieri (singoli o doppi) presenti nel modello scelto

```python
def createWalkwayToHome(depth,intervals,dz,walkebranchDistances,branchL,double,doubleDistance) :
    """
    Funzione che genera un camminamento articolato
    
    Args:
        depth(float): larghezza del camminamento 
        intervals(int) : intervalli di definizione della curva (precisione)
        dz(float) : altezza del camminamento
        walkebranchDistances(float) : distanza tra i rami del camminamento
        branchL(float) : lunghezza dei rami del camminamento
        double(boolean) : discriminante tra cammino singolo o doppio
        doubleDistance(boolean) : eventuale distanza tra due cammini gemelli (caso double = True)
        
    Returns:
        walkway(oggetto hpc): camminamento generato
    """
    # genero il tronco del camminamento
    junctionDim = branchL*1/3    
    walkwayY = walkebranchDistances[len(walkebranchDistances)-1]
        
    pointsCurveTrunk = [[0,0],[0,walkwayY-junctionDim]]
    curveTrunk = createWalkway (pointsCurveTrunk,depth,intervals,dz,[]) 
    walkway = curveTrunk
    
    # genero le diramazioni del camminamento
    for val in walkebranchDistances :
        pointsCurveBranch = [[junctionDim,val],[branchL,val]]
        curveBranch = createWalkway (pointsCurveBranch,depth,intervals,dz,[]) 
        walkway = STRUCT([walkway,curveBranch])
        
        pointsCurveJunction = [[0,val-junctionDim],[0,val],[junctionDim,val]]
        curveJunction = createWalkway (pointsCurveJunction,depth,intervals,dz,[]) 
        walkway = STRUCT([walkway,curveJunction])
    
    walkway = T(1)(depth/2)(walkway)
    
    #se doppio genero il camminamento simmetrico e lo collego all' originale
    if double :
        walkway = T(1)(doubleDistance/2)(walkway)
        walkway = STRUCT([walkway,S(1)(-1),walkway])
        
        pointsDoubleJunction = [[doubleDistance/2+depth/2,walkwayY-junctionDim],[0,walkwayY],[-doubleDistance/2-depth/2,walkwayY-junctionDim]]
        doubleJunction = createWalkway (pointsDoubleJunction,depth,intervals,dz,[])
        walkway = STRUCT([walkway,doubleJunction])
    
    return walkway

```

Definisco una funzione che dati un insieme di punti, un insieme di celle, una texture ed uno spessore genera una "piastra" di forma qualsiasi opportunatamente "texturizzata" (obiettivo: evitare lo "spreading" indesiderato della texture). La piastra viene frammentata ed a ciascun pezzo (di dimensione adeguata) viene applicata la texture.

```python
def createPath(pathPoints,pathCells,textureFileName,n,texDepth) :
    """
    Funzione che genera una piastra con texture applicata
    
    Args:
        pathPoints(List[List[float]]): i punti di bordo della piastra
        pathCells(List[int]): celle convesse che compongo le piastra
        textureFileName(string): nome del file di texture da applicare
        n(int): numero di frammenti da realizzare lungo la dimensione X
        texDepth(float): altezza desiderata della piastra
        
    Returns:
        textured(oggetto hpc): piastra con texture appliacata
    """
    
    points = pathPoints
    cells = pathCells
    
    #genero la piastra
    
    plan =MKPOL([points,cells,1])
    plan = PROD([plan,QUOTE([texDepth])])

    # ricavo le coordinate massime e minime
    minx = points[0][0]
    maxx = points[0][0]
    miny = points[0][1]
    maxy = points[0][1]

    for val in points :
        if val[0] < minx : minx = val[0]
        if val[0] > maxx : maxx = val[0]
        if val[1] < miny : miny = val[1]
        if val[1] > maxy : maxy = val[1]

    dim = (maxx - minx)/n

    myList = []

    # realizzo i frammenti del parellelepipedo che contiene perfettamente la piastra
    i = minx
    while i < maxx:
        j = miny
        while j < maxy:
            if j+dim < maxy : myList.append(STRUCT([T([1,2])([i,j]),CUBOID([dim,dim,texDepth])]))
            else :  myList.append(STRUCT([T([1,2])([i,j]),CUBOID([dim,maxy-j,texDepth])]))
            j += dim
        i += dim

    textured = CUBOID([0,0,0])

    # applico la texture ai vari frammenti intersecati con la piastra originale
    i = 0
    for val in myList:
        val = TEXTURE(textureFileName)(INTERSECTION([plan,val]))
        textured = STRUCT([textured,val])

    #ritorno la piastra
    return textured
```
